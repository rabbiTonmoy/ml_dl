```python
import pandas as pd 
import numpy as np   
```

```python
df = pd.read_csv('../data/data.csv')
```

```python
df.columns = df.columns.str.lower().str.replace(' ', '_')
```

```python
df
```

```python
strings = list(df.dtypes[df.dtypes == 'object'].index)
```

```python
strings
```

```python
for col in strings:
    df[col] = df[col].str.lower().str.replace(' ', '_')
```
